```python
A = np.random.randn(3,2)
```

```python
Q, R = np.linalg.qr(A)
print(A)
print(Q)
print(R)
```

```python
np.allclose(A, np.dot(Q, R)) # check that A=QR
```

```python
U, s, V = np.linalg.svd(A)
print(U.shape, V.shape, s.shape)
```

```python
S = np.zeros(A.shape)
S[:A.shape[1], :A.shape[1]] = np.diag(s)
np.allclose(A, np.dot(U, np.dot(S, V)))
```

```python
U, s, V = np.linalg.svd(A, full_matrices=False)
print(U.shape, V.shape, s.shape)
```

```python
S = np.diag(s)
np.allclose(A, np.dot(U, np.dot(S, V)))
```

```python
#Given two arrays: a = np.array([1, 2, 3, 4, 5]) and b = np.array([6, 7, 8, 9, 10]), compute the element-wise sum, difference, product, and division of a and b.
```

```python
import numpy as np

# Given arrays
a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 10])

# Element-wise operations
sum_result = a + b
diff_result = a - b
product_result = a * b
division_result = a / b

print("Array a:", a)
print("Array b:", b)
print("\nElement-wise Sum:", sum_result)
print("Element-wise Difference (a - b):", diff_result)
print("Element-wise Product:", product_result)
print("Element-wise Division (a / b):", division_result)
```

```python
#Create a 2x2 subplot layout in Matplotlib and display the following in each subplot:

#A line plot of y = x^2.

#A scatter plot of random points.

#A bar chart of some arbitrary data.

#A histogram of random numbers.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# 1. Line plot of y = x^2
x = np.linspace(-10, 10, 100)
y = x ** 2
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Line Plot: y = x^2")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("y")

# 2. Scatter plot of random points
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
axs[0, 1].scatter(x_scatter, y_scatter, color='green')
axs[0, 1].set_title("Scatter Plot: Random Points")
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("y")

# 3. Bar chart of arbitrary data
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 36]
axs[1, 0].bar(categories, values, color='orange')
axs[1, 0].set_title("Bar Chart: Arbitrary Data")
axs[1, 0].set_xlabel("Category")
axs[1, 0].set_ylabel("Value")

# 4. Histogram of random numbers
random_data = np.random.randn(1000)
axs[1, 1].hist(random_data, bins=30, color='purple')
axs[1, 1].set_title("Histogram: Random Numbers")
axs[1, 1].set_xlabel("Value")
axs[1, 1].set_ylabel("Frequency")

# Adjust layout so titles/labels don't overlap
plt.tight_layout()
plt.show()
```

```python
#Create a 5x5 NumPy array with random values and introduce some NaN (Not a Number) values.

#Use NumPy functions to fill in the NaN values with the column mean of the array.
```

```python
import numpy as np

# Create a 5x5 array with random values
arr = np.random.randint(1, 50, size=(5, 5)).astype(float)

# Introduce some NaN values manually
arr[1, 2] = np.nan
arr[3, 4] = np.nan
arr[0, 0] = np.nan

print("Original array with NaN values:")
print(arr)

# Fill NaN values with column mean
col_mean = np.nanmean(arr, axis=0)

# Find indices where NaN is present and replace with column mean
inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_mean, inds[1])

print("\nArray after filling NaN with column mean:")
print(arr)
```
